```python
import matplotlib.pyplot as plt
import community as community_louvain
import numpy as np
import pandas as pd
```

```python

df = pd.read_csv('revision-data3.csv')
```

```python
import igraph as ig

G1 = ig.Graph.TupleList(df.itertuples(index=False), directed=True, weights=True)

# print(G1)
```

```python
# Partition the graph in Directed Louvain clusters

partition = community_louvain.best_partition(G1, weight='weight')
```

```python
# Detect communities in the directed graph with overlapping communities
```

```python
# Partition the graph using the Louvain algorithm and find the best modularity partition using the CPM algorithm (which is the default) and the RBConfigurationVertexPartition method (which is the default)

import leidenalg as la

partition = la.find_partition(G1, la.RBConfigurationVertexPartition)

# partition = la.find_partition(G1, la.ModularityVertexPartition)
# print(partition)
```

```python
# Print the number of clusters
print(len(partition))

# Print the membership of each cluster
print(partition)

# Print the size of each cluster
print([len(c) for c in partition])

# Print the membership of each node
print(partition.membership)

# Print the modularity of the partition
print(partition.quality())

# Draw the graph
# ig.plot(partition, vertex_color=partition.membership)

# Find members which are common in multiple clusters
print("Common members in clusters: ")
for i in range(len(partition)):
    for j in range(i+1, len(partition)):
        if(len(set(partition[i]) & set(partition[j])) > 0):
            print("Cluster {} and Cluster {} have {} common members".format(i, j, len(set(partition[i]) & set(partition[j]))))
        # print("Cluster {} and Cluster {} have {} common members".format(i, j, len(set(partition[i]) & set(partition[j]))))


```
