```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('output.csv')
df.head()
# df.shape
```

```python
# Selected only first 2 columns
df1 = df.iloc[:, 0:2]
# df = df.drop(df.columns[0], axis=1)
df1.head()
```

```python
# club rows which have same user_id and count the number of unique artical_id

df1 = df1.groupby(['Article_id']).nunique()

# df1 = df1.sort_values(by=['User_id'], ascending=False)



df1.head()
```

```python
# select 0 and 2 columns
df2 = df.iloc[:, [0, 2]]
df2.head()
```

```python
# Artical_id and max value of inlink
df2 = df2.groupby('Article_id').max()
df2.head()

# Max value of artical_id 
df2['inlink'].max()

# Max value of which artical_id
# df2['inlink'].idxmax()
```

```python
df3 = df.iloc[:, [0, 3]]
df3.head()
```

```python
df3 = df3.groupby('Article_id').max()
df3.head()
```

```python
df4 = df.iloc[:, [0, 4]]
df4 = df4.groupby('Article_id').max()
df4.head()
```

```python
df5 = df.iloc[:, [0]]

# Add one more column to df5 named 'count' and set all values to 1

# df5.shape

list1 = [1] * df5.shape[0]

df5['Count'] = list1

# Sum of all rows which have same artical_id
df5 = df5.groupby('Article_id').sum()

df5
```

```python
# merge all the dataframes
df6 = pd.merge(df1, df2, on='Article_id')
df6 = pd.merge(df6, df3, on='Article_id')
df6 = pd.merge(df6, df4, on='Article_id')
df6 = pd.merge(df6, df5, on='Article_id')

# Rename the columns of df6
df6.columns = ['no_author', 'inlink', 'external-outlink', 'comment', 'no_revision']

# drop comment column

df6 = df6.drop('comment', axis=1) 

# add one more column named 'credibility' and set it to sum of 'no_author' and 'no_revision' columns

df6['credibility'] = 0.5*(df6['no_author']) + (0.1*(df6['no_revision'] - df6['no_author'])) 

# max value of credibility
df6['credibility'].max()

df6.head()
```

```python
df7 = df6.drop(['inlink','external-outlink','no_revision', 'no_author'], axis=1)
df7.head()
```

```python
# Export the dataframe to csv file
df7.to_csv('temp_final3.csv')
```
