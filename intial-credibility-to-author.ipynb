```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.head()
# df.shape
```

```python
# Selected only first 2 columns
df1 = df.iloc[:, 0:2]
# df = df.drop(df.columns[0], axis=1)
df1.head()
```

```python
# alternate columns
df1 = df1.iloc[:, ::-1]
df1.head()
```

```python
df1 = df1.groupby(['User_id']).nunique()

df1.reset_index(inplace=True)

## Drop last row
df1 = df1.drop(df1.index[-1])

## Replace Title of Article_id to no_autor
df1 = df1.rename(columns={'Article_id': 'intial_credibility'})
df1 = df1.rename(columns={'User_id': 'user_id'})

df1
```

```python
df1.to_csv('intial-credibility-to-author_data_new.csv')
```
