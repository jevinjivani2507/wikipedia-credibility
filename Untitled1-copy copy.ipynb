```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# remove header

df = pd.read_csv('User-PageCount.csv')
df.max()
```

```python
# plot histogram of the column Target

df.plot(kind='bar', figsize=(20,10))

# remove the legend

plt.legend().set_visible(False)

plt.title('Number of Author per Pages', fontsize=25)

# rotate the labels of the x-axis

plt.xticks(rotation=90)

# size of the labels of the x-axis

plt.xticks(fontsize=1)

# size of the labels of the y-axis

# plt.xlabel('Artile_id')

plt.ylabel('Number of Author', fontsize=15)
plt.xlabel('Article_id', fontsize=15)
plt.ylim(0, 3000)

plt.show()
```
