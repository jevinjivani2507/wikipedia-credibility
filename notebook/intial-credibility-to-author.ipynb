```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/data.csv')
df.head()
# df.shape
```

```python
# Selected only first 2 columns
df1 = df.iloc[:, 0:2]
# df = df.drop(df.columns[0], axis=1)
df1.head()
```

```python
# alternate columns
df1 = df1.iloc[:, ::-1]
df1.head()
```

```python
df1 = df1.groupby(['User_id']).nunique()

df1.reset_index(inplace=True)

df1['Article_id'].max()
```

```python
df2 = pd.read_csv('../data/which-author-add-higher-data-perpage.csv', sep=',', header=None, names=['User_id', 'mean of textdata'])
# df2 = df2.iloc[:, ::-1]
df2 = df2.groupby(['User_id']).sum()
df2.reset_index(inplace=True)

df2.head()
```

```python
## merge two dataframes
df3 = pd.merge(df1, df2, on='User_id')

## rename column Article_id to Article_count
df3.rename(columns={'Article_id': 'article_count'}, inplace=True)
df3.rename(columns={'User_id': 'user_id'}, inplace=True)
df3
```

```python
df4 = author_influance = pd.read_csv('../data/author_influance.csv', sep=',', header=None, names=['user_id', 'influance_ppl'])
df4.head()
```

```python
final_credibility = pd.merge(df3, df4, on='user_id')
final_credibility
```

```python
final_credibility['credibility'] = final_credibility['influance_ppl'] + final_credibility['mean of textdata'] + final_credibility['article_count']
final_credibility
```

```python
final_credibility.drop(['influance_ppl', 'mean of textdata', 'article_count'], axis=1, inplace=True)
final_credibility
```

```python
final_credibility.to_csv('../data/intial-credibility-to-author_data_new.csv')
```
