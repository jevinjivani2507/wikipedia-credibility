```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
frequent_page_revised = pd.read_csv('frequent_page_revised.csv', sep=',', header=None, names=['page_id', 'frequency'])

frequent_page_revised['frequency'].max()
```

```python
## bar graph using matplotlib dynamic y axis gap starting from 0 to 50 to 100 to 200 to 400 ans so on
plt.figure(figsize=(20, 10))

## x axis is page id and y axis is frequency

plt.bar(frequent_page_revised['page_id'], frequent_page_revised['frequency'])
plt.xlabel('Page ID')
plt.ylabel('Frequency')
plt.title('Frequent Page Revised')
plt.show()
```

```python
## prefix sum of frequency
frequent_page_revised['prefix_sum'] = frequent_page_revised['frequency'].cumsum()

plt.figure(figsize=(20, 10))
plt.bar(frequent_page_revised['page_id'], frequent_page_revised['prefix_sum'])
plt.xlabel('Page ID')
plt.ylabel('Frequency')
plt.title('Frequent Page Revised')
plt.show()
```
