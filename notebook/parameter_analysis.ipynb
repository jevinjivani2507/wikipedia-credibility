```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
parameter_analysis = pd.read_csv('../data/credibility_withoutUser_Count_.csv', sep=',')
# parameter_analysis = parameter_analysis.drop(columns=['Unnamed: 4', 'Unnamed: 5'])
parameter_analysis
```

```python
parameter_analysis.columns

parameter_analysis['credibility-withoutline'] = parameter_analysis['credibility-withoutline'].astype('float64').apply(lambda x: x*1/parameter_analysis['credibility-withoutline'].max())
parameter_analysis['credibility-whioutDatabyte'] = parameter_analysis['credibility-whioutDatabyte'].astype('float64').apply(lambda x: x*1/parameter_analysis['credibility-whioutDatabyte'].max())
parameter_analysis['credibility-WithoutUsercount'] = parameter_analysis['credibility-WithoutUsercount'].astype('float64').apply(lambda x: x*1/parameter_analysis['credibility-WithoutUsercount'].max())
parameter_analysis['credibility-whioutDatabyte_withMeanData'] = parameter_analysis['credibility-whioutDatabyte_withMeanData'].astype('float64').apply(lambda x: x*1/parameter_analysis['credibility-whioutDatabyte_withMeanData'].max())
parameter_analysis['credibility-withoutline_withMeanData'] = parameter_analysis['credibility-withoutline_withMeanData'].astype('float64').apply(lambda x: x*1/parameter_analysis['credibility-withoutline_withMeanData'].max())
parameter_analysis['HITS'] = parameter_analysis['HITS'].astype('float64').apply(lambda x: x*1/parameter_analysis['HITS'].max())


parameter_analysis['credibility-withoutline'].max()
# parameter_analysis['credibility-whioutDatabyte'].max()
# parameter_analysis['credibility-WithoutUsercount'].max()
```

```python
plt.scatter(parameter_analysis['credibility-whioutDatabyte'], parameter_analysis['credibility-whioutDatabyte'], c='red', label='credibility-whioutDatabyte')
plt.scatter(parameter_analysis['credibility-whioutDatabyte'], parameter_analysis['credibility-WithoutUsercount'], c='blue', label='credibility-WithoutUsercount')
plt.scatter(parameter_analysis['credibility-whioutDatabyte'], parameter_analysis['credibility-withoutline'], c='green', label='credibility-withoutline')
plt.scatter(parameter_analysis['credibility-whioutDatabyte'], parameter_analysis['credibility-whioutDatabyte_withMeanData'], c='yellow', label='credibility-whioutDatabyte_withMeanData')
plt.scatter(parameter_analysis['credibility-whioutDatabyte'], parameter_analysis['credibility-withoutline_withMeanData'], c='black', label='credibility-withoutline_withMeanData')
plt.xlabel('credibility-withoutline')
plt.ylabel('credibility')
plt.legend()
plt.show()
```
