```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# remove header

df = pd.read_csv('revision-data2.csv')
df.head()
```

```python
# get unique values of the column Sources and count it in Target 

PageAuthor = df.groupby('Source')['Target'].nunique()

PageAuthor = pd.DataFrame(PageAuthor)

# max value of the column Target

# Revision['Target'].max()

# sort the values of the column Target descending

PageAuthor = PageAuthor.sort_values(by='Target', ascending=False)

# Remove the first row

PageAuthor = PageAuthor.iloc[4:]

PageAuthor.shape
```

```python
# plot histogram of the column Target

PageAuthor.plot(kind='bar', figsize=(20,10))

# remove the legend

plt.legend().set_visible(False)

plt.title('Number of Author per Pages', fontsize=25)

# rotate the labels of the x-axis

plt.xticks(rotation=90)

# size of the labels of the x-axis

plt.xticks(fontsize=1)

# size of the labels of the y-axis

# plt.xlabel('Artile_id')

plt.ylabel('Number of Author', fontsize=15)
plt.xlabel('Article_id', fontsize=15)
plt.ylim(0, 3000)

plt.show()

# export as PNG

plt.savefig('histogram.png', dpi=300, bbox_inches='tight')
```

```python
# Count number of accurences of each value in the column Target

CountRevision = PageAuthor['Target'].value_counts()

CountRevision = pd.DataFrame(CountRevision)

# sort the values of the column Target descending

CountRevision = CountRevision.sort_values(by='Target', ascending=False)

CountRevision
```

```python
# plot the graph 

CountRevision.plot(kind='bar', figsize=(20, 10))

plt.title('Number of pages per number of author', fontsize=25) 

# rotate the labels of the x-axis

plt.xticks(rotation=90)

# size of the labels of the x-axis

plt.xticks(fontsize=5)

# ylimt of the graph

plt.ylim(0, 200)

# add a title to the histogram
plt.ylabel('Number of pages', fontsize=15) # add y-label
plt.xlabel('Authors Count', fontsize=15) # add x-label

plt.show()
```

```python
df2 = pd.read_csv('revision_data.csv')
df2.head()
```

```python
# get only the column Article_id and External-link

df2 = df2[['Article_id', 'Main']]
df2.head()

df2 = df2.groupby('Article_id')['Main'].max()

# remove the row which has the value of External-link as 0

df2 = df2[df2 != 0]

# max value of the column External-link


# remove the row which has the value of External-link as max value

# sort the values of the column External-link descending

df2 = df2.sort_values(ascending=False)

# remove first 5 rows

df2 = df2[10:]

df2 = pd.DataFrame(df2)

df2.shape
# df2
# df2.head()
```

```python
# plot the graph

df2.plot(kind='bar', figsize=(20, 10))


# x label vertical form
plt.xticks(rotation=90)

# size of the labels of the x-axis

plt.xticks(fontsize=2)

# ylimt of the graph
# plt.xticks(rotation=90) # rotate x-labels
plt.xlabel('Article_id', fontsize=15) # add to x-label to the plot
# adjust y range to see the graph clearly
plt.ylim(0, 2000)
plt.ylabel('Innner-Link', fontsize=15) # add y-label to the plot
plt.title('Incoming Link in Page', fontsize=25) # add title to the plot


plt.show()
```
