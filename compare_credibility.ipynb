```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_hits_page = pd.read_csv('hits_page_credibility_data.csv', sep=',', header=None, names=['page_id', 'credibility'])
df_ours_pahe = pd.read_csv('our_page_credibility_data.csv', sep=',', header=None, names=['page_id', 'credibility'])

# Join the two dataframes on page_id
df = pd.merge(df_hits_page, df_ours_pahe, on='page_id')

# Plot the credibility of the two methods with respect to each other
plt.scatter(df['credibility_x'], df['credibility_y'], c='#6A00AB', alpha=0.5)
plt.xlabel('HITS Credibility')
plt.ylabel('Our Credibility')
plt.title('Credibility Comparison (Pages)')
plt.show()
```
