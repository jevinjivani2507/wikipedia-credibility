```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from cdlib import algorithms
```

```python
df = pd.read_csv('page-page.csv')
df.head()
df.shape
```

```python
import community as community_louvain
import matplotlib.cm as cm
import seaborn as sns
import networkx as nx
```

```python
G1 = nx.read_edgelist('page-page.csv',delimiter=",")
print(G1)
```

```python
# Find communities in graph using Louvain method
partition = community_louvain.best_partition(G1)
num_communities = max(partition.values()) + 1
num_communities


```

```python
#element in every cluster
for i in range(num_communities):
    print("Cluster ",i,": ",list(partition.keys())[list(partition.values()).index(i)])
```

```python
# print all element in community
for i in range(num_communities):
    print("Cluster ",i,": ",[k for k,v in partition.items() if v==i])

    
```

```python
#find overlapping node in every cluster and store in array
overlapping_node = []
for i in range(num_communities):
    for j in range(i+1,num_communities):
        overlapping_node.append(list(set(list(partition.keys())[list(partition.values()).index(i)]) & set(list(partition.keys())[list(partition.values()).index(j)])))
print(overlapping_node)



```

```python
#find overlapping node in every cluster and store in array
# overlapping_node = [][]
for i in range(num_communities):
    print("Cluster ",i,": ",list(partition.keys())[list(partition.values()).index(i)])
    for j in range(num_communities):
        if i!=j:
            print("Cluster ",j,": ",list(partition.keys())[list(partition.values()).index(j)])
            print(set(G1.neighbors(list(partition.keys())[list(partition.values()).index(i)])) & set(G1.neighbors(list(partition.keys())[list(partition.values()).index(j)])))
```

```python
import igraph as ig
import leidenalg as la

G2 = ig.Graph.TupleList(df.itertuples(index=False), directed=True, weights=True)

# print(G2)
```

```python
# Partition the graph using the Leiden algorithm
partition = la.find_partition(G2, la.ModularityVertexPartition)
print(partition)

# store in .txt file
with open('leiden.txt', 'w') as f:
    for item in partition:
        f.write("%s" % item)
    
    
```

```python
# Print the number of clusters
print(len(partition))

# Print the membership of each cluster
print(partition)
f=open("page_leiden.txt","w")
f.write(str(partition))
f.close()
# Print the size of each cluster
print([len(c) for c in partition])

# Print the membership of each node and store it into file
print(partition.membership)

# Find members which are common in multiple clusters
print("Common members in clusters: ")
# for i in range(len(partition)):
#     for j in range(i+1, len(partition)):
#         if(len(set(partition[i]) & set(partition[j])) > 0):
#             # store in file

#             print("Cluster ", i, " and Cluster ", j, " have common members: ", set(partition[i]) & set(partition[j]))

```

```python
coms = algorithms.surprise_communities(G1)
print(coms)
```

```python
# print community
for i in range(len(coms.communities)):
    print("Community ",i,": ",coms.communities[i])

# print overlapping nodes
for i in range(len(coms.communities)):
    for j in range(i+1,len(coms.communities)):
        print("Community ",i,": ",coms.communities[i])
        print("Community ",j,": ",coms.communities[j])
        print("Overlapping nodes: ",set(coms.communities[i]) & set(coms.communities[j]))


#  store it in .txt
with open('page_surprise.txt', 'w') as f:
    for item in coms.communities:
        f.write("%s" % item)
```
