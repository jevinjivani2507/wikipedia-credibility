```python
import pandas as pd
import numpy
```

```python
file_author = open("author_leiden.txt", "r")
dict_author = {}
array_author = file_author.readlines()
c = -1
for i in range(len(array_author)):
  t = array_author[i].split(",")
  for j in range(len(t)):
    if t[j][0] == "[":
      c += 1
      dict_author[c] = []
      dict_author[c].append(int(t[0].split(" ")[1]))
    elif t[j] == "\n":
      continue
    else:
      dict_author[c].append(int(t[j]))

for i in dict_author:
  print(i, dict_author[i])
```

```python
file_page = open("page_leiden.txt", "r")
dict_page = {}
array_page = file_page.readlines()
array_page.pop(0)
c = -1
for i in range(len(array_page)):
  t = array_page[i].split(",")
  for j in range(len(t)):
    if t[j][0] == "[":
      c += 1
      dict_page[c] = []
      dict_page[c].append(int(t[0].split(" ")[1]))
    elif t[j] == "\n":
      continue
    else:
      dict_page[c].append(int(t[j]))

for i in dict_page:
  print(i, dict_page[i])
```

```python
auth = output["Article_id"]
print(type(auth[2018]) == numpy.int64)
```

```python
auth_page = {}
for i in dict_author:
  temp = dict_author[i]
  # temp_arr = []
  for j in temp:
    # if j in auth:
    # print(j)
    # print(type(auth[int(j)]) == numpy.int64)
    if type(auth[int(j)]) == numpy.int64:
      auth_page[int(j)] = [auth[int(j)],]
    else:
      auth_page[int(j)] = auth[int(j)].values
print(auth_page)
```

```python
authors_page_comm = {}
for ii in auth_page:
  comm_page = []
  for i in auth_page[ii]:
    for j in dict_page:
      if i in dict_page[j] and j not in comm_page:
        comm_page.append(j)
        break
  authors_page_comm[ii] = comm_page
print(authors_page_comm)
```

```python
final_dict = {}
for i in authors_page_comm:
  com = -1
  for j in dict_author:
    if i in dict_author[j]:
      com = j
      break
  if com in final_dict:
    for j in authors_page_comm[i]:
      if j not in final_dict[com]:
        final_dict[com].append(j)
  else:
    final_dict[com] = authors_page_comm[i]
print(final_dict)
```
