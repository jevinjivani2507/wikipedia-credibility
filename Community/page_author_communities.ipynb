```python
import pandas as pd
import numpy
```

```python
f_a = open("author_leiden.txt", "r")
d_a = {}
a_a = f_a.readlines()
c = -1
for i in range(len(a_a)):
  t = a_a[i].split(",")
  for j in range(len(t)):
    if t[j][0] == "[":
      c += 1
      d_a[c] = []
      d_a[c].append(int(t[0].split(" ")[1]))
    elif t[j] == "\n":
      # print(t)
      continue
    else:
      d_a[c].append(int(t[j]))

for i in d_a:
  print(i, d_a[i])
```

```python
f_p = open("page_lovian.txt", "r")
d_p = {}
a_p = f_p.readlines()
a_p.pop(0)
c = -1
for i in range(len(a_p)):
  t = a_p[i].split(",")
  for j in range(len(t)):
    if t[j][0] == "[":
      c += 1
      d_p[c] = []
      d_p[c].append(int(t[0].split(" ")[1]))
    elif t[j] == "\n":
      continue
    else:
      d_p[c].append(int(t[j]))

for i in d_p:
  print(i, d_p[i])
```
